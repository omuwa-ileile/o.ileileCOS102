```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("AB_NYC_2019.csv")
data
data.isna().sum()
```

```python
data.shape
data.info()

data.fillna({
    "name": "John Doe",
    "host_name": "Jane Doe",
    "last_review": "No Review",
    "reviews_per_month": 0
}, inplace=True)
data.drop_duplicates(inplace=True)

data.isna().sum()
```

```python
room_price = data.groupby("room_type")["price"].mean()
room_price.plot(kind = "bar")

plt.title("Average Price by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Average Price")

plt.show()
```

```python
neighbourhood_count = data["neighbourhood_group"].value_counts()
neighbourhood_count.plot(
    kind = "pie",
    autopct = "%1.1f%%",
    startangle = 90,
    labels=None 
    )
plt.legend(neighbourhood_count.index,
           title="Neighbourhood_Group", 
           loc="lower left")
plt.title("Number of Listings per Neighbourhood Group")
plt.ylabel("")
plt.show()
```

```python
data.boxplot(column="availability_365", by="room_type")

plt.title("Availability by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Availability (days per year)")

plt.show()
```
