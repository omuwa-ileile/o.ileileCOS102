```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("titanic.csv")
data
data.isna().sum()
```

```python
data.shape
data.info()
```

```python
#data.drop(columns=['Cabin'], inplace=True) #already dropped
data.fillna({
    'Age': data['Age'].median(),
    'Embarked': data['Embarked'].mode()[0]
}, inplace=True)

data['Fare'] = pd.to_numeric(data['Fare'], errors='coerce')
print(data.columns)
data.info()
```

```python
data.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.show()
```

```python
survived = data["Survived"].value_counts()
survived.plot(
    kind = "pie",
    autopct = "%1.1f%%",
    startangle = 90, 
    )
plt.title("Survival Distribution")
plt.ylabel("")
plt.show()
```

```python
data['Age'].plot(kind='hist', bins=8)
plt.title("Age Distribution")
plt.show()
```
