```python
import tkinter as tk
from tkinter import ttk

class Zenith:
    def __init__(self, name):
        self.name = name

    def mutual_services(self):
        return [
            "Lines of Credit",
            "Investment Management and Accounts",
            "Insurance"
        ]

    def unique_services(self):
        return []

class RetailBanking(Zenith):
    def unique_services(self):
        return [
            "Retirement and Education Accounts",
            "Loans and Mortgages",
            "Checking and Savings"
        ]


class GlobalBanking(Zenith):
    def unique_services(self):
        return [
            "Multi-currency Management Services",
            "Foreign Currency Accounts",
            "Foreign Currency Credit Cards",
            "Transborder Advisory Services",
            "Liquidity Management"
        ]


class CommercialBanking(Zenith):
    def unique_services(self):
        return [
            "Advisory Services"
        ]

def get_employee(name, division):
    if division == "Retail Banking":
        return RetailBanking(name)
    elif division == "Global Banking":
        return GlobalBanking(name)
    elif division == "Commercial Banking":
        return CommercialBanking(name)


def show_services():
    name = name_entry.get()
    division = division_combo.get()

    employee = get_employee(name, division)

    if not employee:
        result_label.config(text="Please select a valid division.")
        return

    services = employee.mutual_services() + employee.unique_services()

    output = f"Employee: {employee.name}\nDivision: {division}\n\nServices:\n"
    output += "\n".join(f"- {s}" for s in services)

    result_label.config(text=output)


root = tk.Tk()
root.title("Zenith Bank Service System")
root.geometry("500x400")

# Name input
tk.Label(root, text="Employee Name:").pack()
name_entry = tk.Entry(root, width=40)
name_entry.pack()


tk.Label(root, text="Select Division:").pack()
division_combo = ttk.Combobox(root, values=[
    "Retail Banking",
    "Global Banking",
    "Commercial Banking"
])
division_combo.pack()


tk.Button(root, text="Show Services", command=show_services).pack(pady=10)

result_label = tk.Label(root, text="", justify="left", wraplength=450)
result_label.pack(pady=10)

root.mainloop()
```
