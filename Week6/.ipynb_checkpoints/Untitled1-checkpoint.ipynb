```python
import tkinter as tk
from tkinter import messagebox as msgbox

class deliverycalulator():
    def __init__(self, location, weight):
        self.location = location.lower()
        self.weight = float(weight)
    def calculate_charge(self):
        if self.location == "pau":
            return 2000 if self.weight >= 10 else 1500
        elif self.location == "epe":
            return 5000 if self.weight >= 10 else 4000
        else:
            return None

    def get_charge():
        location = location_entry.get().strip()
        weight = weight_entry.get().strip()

        if not location or not weight:
            msgbox.showerror("Input Error", "Please fill in both fields,")
            return
        try:
            delivery = deliverycalculator(location,weight)
            charge = delivery.calculate_charge()
            if charge is not None:
               messagebox. showinfo ("Delivery Charge", f"The delivery charge is: [charge]")
            else:
                messagebox. Showwarning ("Invalid Location", "Delivery available only to PAU or Epe.")
        except ValueError:
            messagebox. showerror ("Invalid Input", "Please enter a valid number for weight.")

root = tk.Tk()
root.title("Delivery Charge Calculator")
root. geometry("358x200")
tk.Label(root, text="Enter delivery location:").pack(pady=5)
location_entry = tk.Entry(root)
location_entry.pack()
tk. Label(root, text="Enter package weight (kg) :").pack(pady=5)
weight_entry = tk.Entry(root)
weight_entry.pack()
tk.Button(root, text="Calculate Charge", command=get_charge).pack(pady=20) 
root.mainloop()
```
