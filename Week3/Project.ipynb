```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox as msgbox

def submit():
    print("Button clicked!")  

    try:
        df = pd.read_csv("GIG-logistics.csv")

        name = name_entry.get().strip().lower()
        department = department_entry.get().strip().lower()

        print(name, department)  

        match = df[(df["Name"].str.lower() == name.lower()) & (df["Department"].str.lower() == department.lower())]

        if not match.empty:
            welcomeMessage(name, department, df)

            dept_filtered = df[df["Department"].str.contains(department, case=False)]
            print(dept_filtered)

        else:
            msgbox.showerror("Login", "Name or Department entered do not match existing records")

    except Exception as e:
        print("ERROR:", e)

def welcomeMessage(name, department, df):
    window = tk.Toplevel(root)
    window.title("Employee Box")
    window.geometry("500x500")

    label_1 = tk.Label(window, text=f"Welcome {name}")
    label_1.pack()

    label_2 = tk.Label(window, text="Other members in your department:")
    label_2.pack()

    listbox = tk.Listbox(window, width=50)
    listbox.pack()

    dept_members = df[df["Department"].str.lower() == department.lower()]
    dept_members = dept_members[dept_members["Name"].str.lower() != name.lower()]
    
    for index, row in dept_members.iterrows():
        listbox.insert(tk.END, f"{row['Name']} - {row['Department']}")

        
root = tk.Tk()
root.title("Employee Checker")
root.geometry("500x500")

name_label = tk.Label(root, text="Enter your full name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

department_label = tk.Label(root, text="Enter your department:")
department_label.pack()
department_entry = tk.Entry(root)
department_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```
