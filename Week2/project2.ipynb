```python
import numpy as np

def equations():
    print("POLYNOMIAL EQUATION CALCULATOR")
    print("1. Quadratic Equation (Ax^2 + Bx + C = 0)")
    print("2. Cubic Equation (Ax^3 + Bx^2 + Cx + D = 0)")
    print("3. Quartic Equation (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")

    choice = input("Choose the type of equation to solve (1/2/3): ")

    if choice == "1":
        print("\nQUADRATIC EQUATION CALCULATOR")
        a = float(input("Enter value for A: "))
        b = float(input("Enter value for B: "))
        c = float(input("Enter value for C: "))
        roots = np.roots([a, b, c])
        print(f"The roots are: {roots}")

    elif choice == "2":
        print("\nCUBIC EQUATION CALCULATOR")
        a = float(input("Enter value for A: "))
        b = float(input("Enter value for B: "))
        c = float(input("Enter value for C: "))
        d = float(input("Enter value for D: "))
        roots = np.roots([a, b, c, d])
        print(f"The roots are: {roots}")

    elif choice == "3":
        print("\nQUARTIC EQUATION CALCULATOR")
        a = float(input("Enter value for A: "))
        b = float(input("Enter value for B: "))
        c = float(input("Enter value for C: "))
        d = float(input("Enter value for D: "))
        e = float(input("Enter value for E: "))
        roots = np.roots([a, b, c, d, e])
        print(f"The roots are: {roots}")

    else:
        print("INVALID CHOICE.")

equations()
```
