CREATE LIST OF GIRLS NAMES, AGES, HEIGHTS AND SCORES
CREATE LIST OF BOYS NAMES, AGES, HEIGHTS AND SCORES
CREATE STUDENTS TABLE WITH COLUMNS: NAME,AGE,HEIGHT AND SCORE
INSERT DETAILS OF EACH GIRL TO THE STUDENT TABLE
INSERT DETAILS OF EACH BOY TO THE STUDENT TABLE
PRINT STUDENTS TABLE

END

```python
import pandas as pd

girls_names = ["Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola","Aisha","Latifa"]
girls_ages = [17,16,17,18,16,18,17,20,19,17]
girls_heights = [5.5,6.0,5.4,5.9,5.6,5.5,6.1,6.0,5.7,5.5]
girls_scores = [80,85,70,60,76,66,87,95,50,49]

boys_names = ["Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"]
boys_ages = [19,16,18,17,20,19,16,18,17,19]
boys_heights = [5.7,5.9,5.8,6.1,5.9,5.5,6.1,5.4,5.8,5.7]
boys_scores = [74,87,75,68,66,78,87,98,54,60]

girls_df = pd.DataFrame({
    "Name": girls_names,
    "Age": girls_ages,
    "Height": girls_heights,
    "Scores": girls_scores,
})

boys_df = pd.DataFrame({
    "Name": boys_names,
    "Age": boys_ages,
    "Height": boys_heights,
    "Scores": boys_scores,
})
students_df = pd.concat([girls_df, boys_df])
print(students_df.to_string(index=False))
```
